```julia
using BifurcationKit, Plots
```

```julia
F(x, p) =  @. p[1] + x^2
```

```julia

prob = BifurcationProblem(F, [-2.], [-1.], 1;
    record_from_solution = (x,p; k...) -> x[1])
```

```julia
br = continuation(prob, PALC(), ContinuationPar(p_min = -1., p_max = 1.))
plot(br)
```
