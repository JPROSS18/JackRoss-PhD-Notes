```python
from juliacall import Main as jl

# Load the package
jl.seval("using BifurcationKit")

# Example: define a simple 1D system and continuation setup
jl.seval("""
f(u, p) = p * u - u^3  # normal form
using BifurcationKit

prob = BifurcationProblem(f, 0.1, 0.0, (@lens _.p))
opts = ContinuationPar(ds=0.01, dsmax=0.1, pMax=2.0)

br = continuation(prob, PALC(), opts)
""")

# Access results
print(jl.seval("length(br)"))
```
